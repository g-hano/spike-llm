```python
from datasets import load_dataset

ds = load_dataset("garage-bAInd/Open-Platypus")
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('Chan-Y/bpe-32768')
```

```python
ds
```

```python
def filter_by_token_length(example):
    input_tokens = tokenizer(example['input'])['input_ids']
    output_tokens = tokenizer(example['output'])['input_ids']
    return len(input_tokens) <= 200 and len(output_tokens) <= 150

ds['train'] = ds['train'].filter(filter_by_token_length)

display(ds)
```

```python
from datasets import load_dataset

ds = load_dataset("openai/gsm8k", "main")
```

```python
def filter_by_token_length(example):
    input_tokens = tokenizer(example['question'])['input_ids']
    output_tokens = tokenizer(example['answer'])['input_ids']
    return len(input_tokens) <= 200 and len(output_tokens) <= 150

ds['train'] = ds['train'].filter(filter_by_token_length)
ds['test'] = ds['test'].filter(filter_by_token_length)

display(ds)
```

```python
ds['train'].save_to_disk('gsm8k-train')
ds['test'].save_to_disk('gsm8k-test')
!zip -r gsm8k-train.zip gsm8k-train
!zip -r gsm8k-test.zip gsm8k-test
```

```python
from datasets import load_from_disk

platy = load_from_disk('/content/Open-Platypus-i200-o150')
platy
```

```python
gtest = load_from_disk('gsm8k-test')
gtrain = load_from_disk('gsm8k-train')
gtrain[0]
```

```python
from datasets import concatenate_datasets

gsm8k_ds = concatenate_datasets([gtrain, gtest])

gsm8k_ds = gsm8k_ds.rename_column('question', 'input')
gsm8k_ds = gsm8k_ds.rename_column('answer', 'output')

display(gsm8k_ds)
```

```python
orca = load_from_disk('orca-math-i200-o150')
orca
```

```python
orca = orca.rename_column('question', 'input')
orca = orca.rename_column('answer', 'output')
orca
```

```python
dataset = concatenate_datasets([platy, gsm8k_ds, orca])
dataset
```

```python
dataset.num_rows == platy.num_rows +gsm8k_ds.num_rows +orca.num_rows
```

```python
dataset.save_to_disk('merged-math')
```
